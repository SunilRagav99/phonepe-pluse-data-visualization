```python
import os
import pandas as pd

paath=r"C:\Users\sunil\OneDrive\Desktop\phone\pulse\data\aggregated\transaction\country\india\state"
arg_trans=os.listdir(paath)

column={'State': [], 'Year': [], 'Quarter': [], 'Transaction_type': [], 'Transaction_count': [],
            'Transaction_amount': []}

for state in arg_trans:
    current_state=paath + "/" + state
    agg_trans_list=os.listdir(current_state)
    
    for year in agg_trans_list:
        current_year=current_state + "/" + year
        agg_trans_year=os.listdir(current_year)
        # print(agg_trans_list)
        # print(agg_trans_year)
        for file in agg_trans_year:
            current_data=current_year+"/"+ file
            agg_trans_data=open(current_data,'r')
            data=json.load(agg_trans_data)
            for a in data["data"]["transactionData"]:
                name=a["name"]
                amount=a['paymentInstruments'][0]["amount"]
                count=a['paymentInstruments'][0]["count"]
                column["State"].append(state)
                column["Year"].append(year)
                column["Quarter"].append(file.strip(".json"))
                column["Transaction_type"].append(name)
                column['Transaction_count'].append(count)
                column["Transaction_amount"].append(amount)
arg_trans_df=pd.DataFrame(column)
arg_trans_df
```

```python
import os
import pandas as pd

paath1=r"C:\Users\sunil\OneDrive\Desktop\phone\pulse\data\aggregated\insurance\country\india\state"
arg_insure=os.listdir(paath1)

column1={'State': [], 'Year': [], 'Quarter': [], 'type': [], 'Insurence_count': [],
            'Insurence_amount': []}

for state in arg_insure:
    current_state=paath + "/" + state
    agg_insure_list=os.listdir(current_state)
    
    for year in agg_insure_list:
        current_year=current_state + "/" + year
        agg_insure_year=os.listdir(current_year)
        # print(agg_trans_list)
        # print(agg_trans_year)
        for file in agg_insure_year:
            current_data=current_year+"/"+ file
            agg_insure_data=open(current_data,'r')
            data=json.load(agg_insure_data)
            # print(data)
            for a in data["data"]["transactionData"]:
                name=a["name"]
                amount=a['paymentInstruments'][0]["amount"]
                count=a['paymentInstruments'][0]["count"]
                column1["State"].append(state)
                column1["Year"].append(year)
                column1["Quarter"].append(file.strip(".json"))
                column1["type"].append(name)
                column1['Insurence_count'].append(count)
                column1["Insurence_amount"].append(amount)
arg_insurence_df=pd.DataFrame(column1)
arg_insurence_df
```

```python
import os
import pandas
paath2=r"C:\Users\sunil\OneDrive\Desktop\phone\pulse\data\aggregated\user\country\india\state"
arg_user=os.listdir(paath2)

column2={'State': [], 'Year': [], 'Quarter': [], 'Brand': [], 'count': [],
            'percentage': []}

for state in arg_user:
    current_state=paath2 + "/" + state
    agg_user_list=os.listdir(current_state)
    # print(agg_user_list)
    
    for year in agg_user_list:
        current_year=current_state + "/" + year
        agg_user_year=os.listdir(current_year)
        # print(agg_user_list)
        # print(agg_user_year)
        for file in agg_user_year:
            current_data=current_year+"/"+ file
            agg_user_data=open(current_data,'r')
            data1=json.load(agg_user_data)
            # print(data1["data"]["usersByDevice"])
            try:
                for d in data1["data"]["usersByDevice"]:
                    name=d["brand"]
                    amount=d['percentage']
                    count=d['count']
                    column2["State"].append(state)
                    column2["Year"].append(year)
                    column2["Quarter"].append(file.strip(".json"))
                    column2["Brand"].append(name)
                    column2['count'].append(count)
                    column2["percentage"].append(amount)
            except:
                pass
arg_user_df=pd.DataFrame(column2)
arg_user_df
```

```python
path3 = r"C:\Users\sunil\OneDrive\Desktop\phone\pulse\data\map\transaction\hover\country\india\state"

map_trans_list = os.listdir(path3)

columns3 = {'State': [], 'Year': [], 'Quarter': [], 'District': [], 'Count': [],
            'Amount': []}

for state in map_trans_list:
    cur_state = path3 + "/" + state
    map_year_list = os.listdir(cur_state)
    
    for year in map_year_list:
        cur_year = cur_state  + "/"+ year
        map_file_list = os.listdir(cur_year)
        
        for file in map_file_list:
            cur_file = cur_year +"/"+ file
            data = open(cur_file, 'r')
            C = json.load(data)
            # print(C)
            
            for i in C["data"]["hoverDataList"]:
                district = i["name"]
                count = i["metric"][0]["count"]
                amount = i["metric"][0]["amount"]
                columns3["District"].append(district)
                columns3["Count"].append(count)
                columns3["Amount"].append(amount)
                columns3['State'].append(state)
                columns3['Year'].append(year)
                columns3['Quarter'].append(file.strip('.json'))
                
df_map_trans = pd.DataFrame(columns3)
df_map_trans
```

```python
path4 = r"C:\Users\sunil\OneDrive\Desktop\phone\pulse\data\map\user\hover\country\india\state"

map_user_list = os.listdir(path4)

columns4 = {"State": [], "Year": [], "Quarter": [], "District": [],
            "RegisteredUser": [], "AppOpens": []}


for state in map_user_list:
    cur_state = path4 + "/" + state
    map_year_list = os.listdir(cur_state)
    
    for year in map_year_list:
        cur_year = cur_state  + "/"+ year
        map_file_list = os.listdir(cur_year)
        
        for file in map_file_list:
            cur_file = cur_year +"/"+ file
            data = open(cur_file, 'r')
            C = json.load(data)
            # print(C)
            
            for i in C["data"]["hoverData"].items():
                district = i[0]
                registereduser = i[1]["registeredUsers"]
                appOpens = i[1]['appOpens']
                columns4["District"].append(district)
                columns4["RegisteredUser"].append(registereduser)
                columns4["AppOpens"].append(appOpens)
                columns4['State'].append(state)
                columns4['Year'].append(year)
                columns4['Quarter'].append(file.strip('.json'))
df_map_user = pd.DataFrame(columns4)
df_map_user
```

```python
path5 = r"C:\Users\sunil\OneDrive\Desktop\phone\pulse\data\top\transaction\country\india\state"

top_trans_list = os.listdir(path5)
columns5 = {'State': [], 'Year': [], 'Quarter': [], 'Pincode': [], 'Transaction_count': [],
            'Transaction_amount': []}

for state in top_trans_list:
    cur_state = path5 + "/" + state
    top_year_list = os.listdir(cur_state)
    
    for year in top_year_list:
        cur_year = cur_state  + "/"+ year
        top_file_list = os.listdir(cur_year)
        
        for file in top_file_list:
            cur_file = cur_year +"/"+ file
            data = open(cur_file, 'r')
            E = json.load(data)
            
            for i in E['data']['pincodes']:
                name = i['entityName']
                count = i['metric']['count']
                amount = i['metric']['amount']
                columns5['Pincode'].append(name)
                columns5['Transaction_count'].append(count)
                columns5['Transaction_amount'].append(amount)
                columns5['State'].append(state)
                columns5['Year'].append(year)
                columns5['Quarter'].append(file.strip('.json'))
df_top_trans=pd.DataFrame(columns5)
df_top_trans
```

```python
path6 = r"C:\Users\sunil\OneDrive\Desktop\phone\pulse\data\top\user\country\india\state"
top_user_list = os.listdir(path6)
columns6 = {'State': [], 'Year': [], 'Quarter': [], 'Pincode': [],
            'RegisteredUsers': []}

for state in top_user_list:
    cur_state = path6  + "/"+ state
    top_year_list = os.listdir(cur_state)
    
    for year in top_year_list:
        cur_year = cur_state  + "/"+ year
        top_file_list = os.listdir(cur_year)
        
        for file in top_file_list:
            cur_file = cur_year + "/" +file
            data = open(cur_file, 'r')
            F = json.load(data)
            
            for i in F['data']['pincodes']:
                name = i['name']
                registeredUsers = i['registeredUsers']
                columns6['Pincode'].append(name)
                columns6['RegisteredUsers'].append(registeredUsers)
                columns6['State'].append(state)
                columns6['Year'].append(year)
                columns6['Quarter'].append(file.strip('.json'))
df_top_user = pd.DataFrame(columns6) 
df_top_user
```
